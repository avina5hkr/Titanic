```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, ann_prediction)
cm
```

```python
# Manually calculating accuracy and precision
TP = cm[1][1] #True Positives
TN = cm[0][0] #True Negatives
FP = cm[1][0] #False Positives
FN = cm[0][1] #False Negatives
```

<h4>Accuracy</h4> is as the name goes. It measures the percentage of correct predictions made by the model. 
              <br>Sum all True predictions and divide by the total number of predictions.<br>
<h4>Precision</h4> is the closeness of two or more measurements to each other. 
              <br>Divide True Positives by Total Positives.<br>
<h4>Recall</h4> is the ratio of correctly predicted positive observations to all observations in the actual class. <br>
              Divide True Positives by the sum of True Positives and False Negatives.<br>
<h4>F1 Score</h4> is the weighted average of Precision and Recall. F1 Score might be a better measure to use if I need to seek a balance between Precision &Recall, AND if there is an uneven class distribution. <br>
Multiply Precision and Recall, divide the result by the sum of Precision and Recall, and then multiply the final result by2.

```python
#Accuracy
Accuracy = (TP + TN)/(TP + TN + FN + FP)
Accuracy
```

```python
 #Precision
Precision = TP / (TP + FP)
Precision
```

```python
#Recall
Recall = TP / (TP + FN)
Recall
```

```python
#F1 Score
F1_Score = 2 * Precision * Recall / (Precision + Recall)
F1_Score
```
